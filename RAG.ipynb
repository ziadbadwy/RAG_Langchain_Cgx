```python
import os
import requests
import gradio as gr
from Helper import helper
```

#**Downloading Model**

```python
url = 'https://gpt4all.io/models/gguf/Meta-Llama-3-8B-Instruct.Q4_0.gguf'
filename = "Meta-Llama-3-8B-Instruct.Q4_0.gguf"
response = requests.get(url, stream=True)
with open(filename, 'wb') as file:
    file.write(response.content)
```

```python
#deallocating from memory
del response
```

```python
#get model path
HOME = os.getcwd()
model_path = os.path.join(HOME,filename)
```

#**Load LLM**

```python
# Load the helper class
hp = helper("https://aiindex.stanford.edu/wp-content/uploads/2024/04/HAI_AI-Index-Report-2024.pdf")
```

```python
#load llm
llm = hp.build_agent(verbose=False,streaming=True,max_tokens=50,gpt4all_path=model_path)
```

```python
#generate response
hp.llm_response("what do you know about world war II",llm)
```

#**Build GUI**

```python
def respond(message, chat_history):
    bot_message = hp.llm_response(message,llm)
    return bot_message
# Define the interface and launch the app
gr.ChatInterface(respond).launch()
```

#**Evaluation Pipeline**

```python
score = hp.eval_pipeline(llm,'questions.txt')
```

```python
print(f'score is : {score}')
```
